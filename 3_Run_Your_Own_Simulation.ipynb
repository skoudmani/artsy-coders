# A Mini Solar System ☀️🌍

Run the cell below. It places the **Sun**, **Venus** and **Earth**, then plays time forward so you can watch them orbit. A dot shows each body's current position and a line traces the path it has travelled.

Building the animation can take a few seconds or even minutes! When the little video player appears, press ▶.


````python
# Self-contained: imports everything and defines its own simulation.
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
%matplotlib inline
plt.rcParams['animation.embed_limit'] = 50

# === Settings you can play with ===
dt = 60 * 60 * 24 * 5   # how much time passes each step (here: 5 days)
steps = 150             # how many steps to play

# === The bodies: name, mass, starting position, starting speed, colour ===
bodies = [
    {'name': 'Sun',   'mass': 1.989e30, 'pos': np.array([0.0, 0.0]),       'vel': np.array([0.0, 0.0]),     'color': 'orange'},
    {'name': 'Venus', 'mass': 4.867e24, 'pos': np.array([1.082e11, 0.0]),  'vel': np.array([0.0, 35.02e3]), 'color': 'goldenrod'},
    {'name': 'Earth', 'mass': 5.972e24, 'pos': np.array([1.496e11, 0.0]),  'vel': np.array([0.0, 29.78e3]), 'color': 'blue'},
]

G = 6.67430e-11   # the strength of gravity (a fixed number of nature)
for body in bodies:
    body['trajectory'] = [body['pos'].copy()]

def step():
    """Move every body forward by one time-step under gravity."""
    forces = [np.zeros(2) for _ in bodies]
    for i, b1 in enumerate(bodies):
        for j, b2 in enumerate(bodies):
            if i == j:
                continue
            r_vec = b2['pos'] - b1['pos']
            r = np.linalg.norm(r_vec)
            if r == 0:
                continue
            forces[i] += G * b1['mass'] * b2['mass'] * r_vec / r**3
    for i, body in enumerate(bodies):
        body['vel'] += (forces[i] / body['mass']) * dt
        body['pos'] += body['vel'] * dt
        body['trajectory'].append(body['pos'].copy())

# === Draw it ===
fig, ax = plt.subplots(figsize=(6, 6))
ax.set_xlim(-2e11, 2e11)
ax.set_ylim(-2e11, 2e11)
ax.set_aspect('equal')
ax.set_title("My Solar System")
ax.set_xlabel("X (m)")
ax.set_ylabel("Y (m)")

lines = [ax.plot([], [], '-', lw=1, color=b['color'])[0] for b in bodies]
dots  = [ax.plot([], [], 'o', color=b['color'])[0] for b in bodies]

def update(frame):
    step()
    for i, body in enumerate(bodies):
        traj = np.array(body['trajectory'])
        lines[i].set_data(traj[:, 0], traj[:, 1])
        dots[i].set_data([traj[-1, 0]], [traj[-1, 1]])
    return lines + dots

ani = FuncAnimation(fig, update, frames=steps, interval=30, blit=False)
plt.close(fig)
HTML(ani.to_jshtml())
````

# ⭐ Bonus: Tiny Changes, Chaotic Results

Some systems are **super sensitive**: change the start by a hair and the ending is completely different. Run the cell below to watch two heavy stars and one small planet.

Then re-run it after changing the planet's starting position from `2.0e10` to **`2.01e10`**: a difference of less than 1%. Watch how differently it ends up! 🌪️


````python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
%matplotlib inline
plt.rcParams['animation.embed_limit'] = 50

dt = 60 * 60 * 12   # half a day per step
steps = 600

bodies = [
    {'name': 'Star A',   'mass': 5.0e26, 'pos': np.array([-1.0e10, 0.0]), 'vel': np.array([0.0, -1000.0]), 'color': 'cyan'},
    {'name': 'Star B',   'mass': 5.0e26, 'pos': np.array([ 1.0e10, 0.0]), 'vel': np.array([0.0,  1000.0]), 'color': 'magenta'},
    {'name': 'Planet C', 'mass': 1.0e22, 'pos': np.array([0.0, 2.0e10]),  'vel': np.array([500.0, 0.0]),   'color': 'yellow'},
    # 👆 try changing 2.0e10 to 2.01e10 and re-running!
]

G = 6.67430e-11
for body in bodies:
    body['trajectory'] = [body['pos'].copy()]

def step():
    forces = [np.zeros(2) for _ in bodies]
    for i, b1 in enumerate(bodies):
        for j, b2 in enumerate(bodies):
            if i == j:
                continue
            r_vec = b2['pos'] - b1['pos']
            r = np.linalg.norm(r_vec)
            if r == 0:
                continue
            forces[i] += G * b1['mass'] * b2['mass'] * r_vec / r**3
    for i, body in enumerate(bodies):
        body['vel'] += (forces[i] / body['mass']) * dt
        body['pos'] += body['vel'] * dt
        body['trajectory'].append(body['pos'].copy())

fig, ax = plt.subplots(figsize=(8, 8))
ax.set_xlim(-5e10, 5e10)
ax.set_ylim(-5e10, 5e10)
ax.set_aspect('equal')
ax.set_title("Tiny Changes, Chaotic Results")
ax.set_xlabel("X (m)")
ax.set_ylabel("Y (m)")

lines = [ax.plot([], [], '-', lw=1, color=b['color'])[0] for b in bodies]
dots  = [ax.plot([], [], 'o', color=b['color'], markersize=7)[0] for b in bodies]

def update(frame):
    step()
    for i, body in enumerate(bodies):
        traj = np.array(body['trajectory'])
        lines[i].set_data(traj[:, 0], traj[:, 1])
        dots[i].set_data([traj[-1, 0]], [traj[-1, 1]])
    return lines + dots

ani = FuncAnimation(fig, update, frames=steps, interval=20, blit=False)
plt.close(fig)
HTML(ani.to_jshtml())
````

# ⭐ Bonus: Orbiting in 3D

Our Universe is 3D, and so can our simulations be! The cell below sends a satellite looping around a central mass in **three dimensions**: it moves across the screen *and* into and out of it.

🎮 **Experiment:** in the cell, change the satellite's starting speed `np.array([0.0, 1000.0, 500.0])`:
- set the last number to `0.0` → `[0.0, 1000.0, 0.0]`. What shape is the orbit now?
- make it bigger → `[0.0, 1000.0, 1000.0]`. Does the orbit get taller or more twisted?

Or change its starting position `np.array([5.0e9, 0.0, 0.0])` to `[5.0e9, 0.0, 2.0e9]` and see what happens. ✨


````python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D   # enables 3D plotting
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
%matplotlib inline
plt.rcParams['animation.embed_limit'] = 50

dt = 60 * 60 * 24 * 3   # 3 days per step
steps = 200

# Positions and velocities now have THREE numbers: [x, y, z]
bodies = [
    {'name': 'Central Mass', 'mass': 1.0e26, 'pos': np.array([0.0, 0.0, 0.0]),  'vel': np.array([0.0, 0.0, 0.0]),      'color': 'gold'},
    {'name': 'Satellite',    'mass': 1.0e10, 'pos': np.array([5.0e9, 0.0, 0.0]),'vel': np.array([0.0, 1000.0, 500.0]), 'color': 'lime'},
]

G = 6.67430e-11
for body in bodies:
    body['trajectory'] = [body['pos'].copy()]

def step():
    forces = [np.zeros(3) for _ in bodies]
    for i, b1 in enumerate(bodies):
        for j, b2 in enumerate(bodies):
            if i == j:
                continue
            r_vec = b2['pos'] - b1['pos']
            r = np.linalg.norm(r_vec)
            if r == 0:
                continue
            forces[i] += G * b1['mass'] * b2['mass'] * r_vec / r**3
    for i, body in enumerate(bodies):
        body['vel'] += (forces[i] / body['mass']) * dt
        body['pos'] += body['vel'] * dt
        body['trajectory'].append(body['pos'].copy())

fig = plt.figure(figsize=(9, 9))
ax = fig.add_subplot(111, projection='3d')
limit = 6.0e9
ax.set_xlim([-limit, limit]); ax.set_ylim([-limit, limit]); ax.set_zlim([-limit, limit])
ax.set_title("Simple 3D Orbit")
ax.set_xlabel("X (m)"); ax.set_ylabel("Y (m)"); ax.set_zlabel("Z (m)")

lines = [ax.plot([], [], [], '-', lw=1, color=b['color'])[0] for b in bodies]
dots  = [ax.plot([], [], [], 'o', color=b['color'], markersize=7)[0] for b in bodies]

def update(frame):
    step()
    for i, body in enumerate(bodies):
        traj = np.array(body['trajectory'])
        lines[i].set_data(traj[:, 0], traj[:, 1])
        lines[i].set_3d_properties(traj[:, 2])
        dots[i].set_data([traj[-1, 0]], [traj[-1, 1]])
        dots[i].set_3d_properties([traj[-1, 2]])
    return lines + dots

ani = FuncAnimation(fig, update, frames=steps, interval=50, blit=False)
plt.close(fig)
HTML(ani.to_jshtml())
````
