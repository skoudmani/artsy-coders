# Section 1: Meet the Galaxy 🔭

Our galaxy is made of gas cells and star particles, each with a position in 3D space (`x`, `y`, `z`) and a `mass`. We load them into a table using **Pandas**.

Run the cell below to peek at the first few rows of the gas data. Each row is one blob of gas floating in space.


```python
# --- self-contained setup (safe to run this cell on its own) ---
import os
import numpy as np
import pandas as pd

def load_galaxy(component="gas"):
    """Load the galaxy data ('gas' or 'stars').
    If the data file isn't available (e.g. the environment hiccuped),
    we fall back to a built-in synthetic galaxy so the cell still runs."""
    path = f"./data/galaxy/{component}.csv"
    if os.path.exists(path):
        return pd.read_csv(path)
    # Fallback: a small synthetic spiral galaxy centred at (151, 151)
    rng = np.random.default_rng(0 if component == "gas" else 1)
    n = 80000 if component == "gas" else 25000
    r = rng.exponential(2.0, n)
    spread = 0.25 if component == "gas" else 0.5
    theta = 2.5 * r + rng.normal(0, spread, n) + rng.integers(0, 2, n) * np.pi
    x = 151.0 + r * np.cos(theta)
    y = 151.0 + r * np.sin(theta)
    z = 151.0 + rng.normal(0, 0.35, n)
    mass = rng.exponential(5e3, n)
    return pd.DataFrame({"x": x, "y": y, "z": z, "mass": mass})

# Load the gas particles and look at the first few rows
gas_data = load_galaxy("gas")
gas_data.head()
```

Galaxies are famously full of **stars**! Let's load the star particles and lay them **on top of** the gas, in a bright **gold** so they stand out.

Combining gas (black) and stars (gold) on one plot gives a much richer picture of what a galaxy really looks like. 🌟

# Section 4: Turning Gas into a Galaxy Image 🌠

So far each dot is one particle. But what if we want to **see how much gas** sits in each part of the galaxy, like a heatmap? That's closer to how a real telescope builds an image.

We'll:
1. Divide space into a grid (like **pixels**),
2. Add up the gas **mass** in each pixel,
3. Display it with `imshow()`, using a `LogNorm` so both faint and bright regions show up.


```python
# --- self-contained setup (safe to run this cell on its own) ---
import os
import numpy as np
import pandas as pd

def load_galaxy(component="gas"):
    """Load the galaxy data ('gas' or 'stars').
    If the data file isn't available (e.g. the environment hiccuped),
    we fall back to a built-in synthetic galaxy so the cell still runs."""
    path = f"./data/galaxy/{component}.csv"
    if os.path.exists(path):
        return pd.read_csv(path)
    # Fallback: a small synthetic spiral galaxy centred at (151, 151)
    rng = np.random.default_rng(0 if component == "gas" else 1)
    n = 80000 if component == "gas" else 25000
    r = rng.exponential(2.0, n)
    spread = 0.25 if component == "gas" else 0.5
    theta = 2.5 * r + rng.normal(0, spread, n) + rng.integers(0, 2, n) * np.pi
    x = 151.0 + r * np.cos(theta)
    y = 151.0 + r * np.sin(theta)
    z = 151.0 + rng.normal(0, 0.35, n)
    mass = rng.exponential(5e3, n)
    return pd.DataFrame({"x": x, "y": y, "z": z, "mass": mass})

import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
%matplotlib inline

gas_data = load_galaxy("gas")
x = gas_data['x']
y = gas_data['y']
mass = gas_data['mass']

# Set up the grid (300 x 300 pixels)
nbins = 300
x_edges = np.linspace(140, 160, nbins + 1)
y_edges = np.linspace(140, 160, nbins + 1)

# Add up the mass in each pixel
image, _, _ = np.histogram2d(x, y, bins=[x_edges, y_edges], weights=mass)
image[image <= 0] = 100   # small floor so the log colour scale behaves

plt.figure(figsize=(8, 8))
plt.imshow(image.T, origin='lower', extent=[140, 160, 140, 160],
           cmap='inferno', norm=LogNorm())
plt.title("Gas Density Map of the Galaxy")
plt.xlabel("X (kpc)")
plt.ylabel("Y (kpc)")
plt.colorbar(label="Gas Mass (Solar Masses)")
plt.show()
```

🔥 You've made a **density map** where bright areas have lots of gas, dark areas have little!

- Try a different `cmap`, such as `"plasma"`, `"viridis"`, or `"gray"`.

# Section 5: Viewing the Galaxy Edge-On 🌌

Until now we've looked **face-on** (using `x` and `y`), like looking down on a spinning disc. But galaxies are **3D**! Let's turn our view **sideways** and look **edge-on**, using `x` and `z` instead.

This reveals vertical features such as gas puffed up **above and below** the disc.


```python
# --- self-contained setup (safe to run this cell on its own) ---
import os
import numpy as np
import pandas as pd

def load_galaxy(component="gas"):
    """Load the galaxy data ('gas' or 'stars').
    If the data file isn't available (e.g. the environment hiccuped),
    we fall back to a built-in synthetic galaxy so the cell still runs."""
    path = f"./data/galaxy/{component}.csv"
    if os.path.exists(path):
        return pd.read_csv(path)
    # Fallback: a small synthetic spiral galaxy centred at (151, 151)
    rng = np.random.default_rng(0 if component == "gas" else 1)
    n = 80000 if component == "gas" else 25000
    r = rng.exponential(2.0, n)
    spread = 0.25 if component == "gas" else 0.5
    theta = 2.5 * r + rng.normal(0, spread, n) + rng.integers(0, 2, n) * np.pi
    x = 151.0 + r * np.cos(theta)
    y = 151.0 + r * np.sin(theta)
    z = 151.0 + rng.normal(0, 0.35, n)
    mass = rng.exponential(5e3, n)
    return pd.DataFrame({"x": x, "y": y, "z": z, "mass": mass})

import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
%matplotlib inline

gas_data = load_galaxy("gas")
x = gas_data['x']
z = gas_data['z']
mass = gas_data['mass']

nbins = 300
x_edges = np.linspace(140, 160, nbins + 1)
z_edges = np.linspace(140, 160, nbins + 1)

image_xz, _, _ = np.histogram2d(x, z, bins=[x_edges, z_edges], weights=mass)
image_xz[image_xz <= 0] = 100

plt.figure(figsize=(10, 8))
plt.imshow(image_xz.T, origin='lower', extent=[140, 160, 130, 170],
           cmap='inferno', norm=LogNorm())
plt.title("Edge-On View of the Galaxy")
plt.xlabel("X (kpc)")
plt.ylabel("Z (kpc)")
plt.colorbar(label="Gas Mass (log scale)")
plt.show()
```

# Section 6: Galaxy Rotation Animation 🌀

Let's bring the galaxy to life! This cell **animates** the galaxy turning from face-on to edge-on, so you can see its 3D shape.

It works by rotating the particle positions a little more each frame and re-drawing them. Press ▶ on the player when it appears (it can take a few seconds to build).


```python
# --- self-contained setup (safe to run this cell on its own) ---
import os
import numpy as np
import pandas as pd

def load_galaxy(component="gas"):
    """Load the galaxy data ('gas' or 'stars').
    If the data file isn't available (e.g. the environment hiccuped),
    we fall back to a built-in synthetic galaxy so the cell still runs."""
    path = f"./data/galaxy/{component}.csv"
    if os.path.exists(path):
        return pd.read_csv(path)
    # Fallback: a small synthetic spiral galaxy centred at (151, 151)
    rng = np.random.default_rng(0 if component == "gas" else 1)
    n = 80000 if component == "gas" else 25000
    r = rng.exponential(2.0, n)
    spread = 0.25 if component == "gas" else 0.5
    theta = 2.5 * r + rng.normal(0, spread, n) + rng.integers(0, 2, n) * np.pi
    x = 151.0 + r * np.cos(theta)
    y = 151.0 + r * np.sin(theta)
    z = 151.0 + rng.normal(0, 0.35, n)
    mass = rng.exponential(5e3, n)
    return pd.DataFrame({"x": x, "y": y, "z": z, "mass": mass})

import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML
%matplotlib inline
plt.rcParams['animation.embed_limit'] = 50

gas_data = load_galaxy("gas")

# Centre the coordinates on the galaxy centre (~151)
centre = 151.0
x_c = gas_data['x'] - centre
y_c = gas_data['y'] - centre
z_c = gas_data['z'] - centre

fig, ax = plt.subplots(figsize=(8, 8))
scat = ax.scatter(x_c, y_c, s=0.01, color='black')
ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10)
ax.set_title("Rotating Galaxy")
ax.set_xlabel("X (kpc)")

def update(frame):
    angle = np.radians(frame)
    cos_a, sin_a = np.cos(angle), np.sin(angle)
    # Rotate around the X-axis: mix the Y and Z coordinates
    y_rot = y_c * cos_a - z_c * sin_a
    scat.set_offsets(np.c_[x_c, y_rot])
    ax.set_ylabel(f"View angle: {frame}°")
    return scat,

ani = animation.FuncAnimation(fig, update, frames=np.arange(0, 180, 2),
                              interval=100, blit=False)
plt.close(fig)
HTML(ani.to_jshtml())
```

## 🌌 The Challenge
Astronomers study **supermassive black hole binaries** which are two giant black holes orbiting each other, wrapped in swirling gas. The next cell loads the data for one of these systems. **The cell after that is yours to play with**: visualise it however you like, using the tricks from above.


```python
# --- self-contained setup (safe to run this cell on its own) ---
import os
import numpy as np
import pandas as pd

def load_smbh(component="smbhs"):
    """Load the black-hole binary data ('smbhs' or 'gas').
    Falls back to a built-in synthetic system if the file isn't available."""
    path = f"./data/smbhs/{component}.csv"
    if os.path.exists(path):
        return pd.read_csv(path)
    if component == "smbhs":
        # Two black holes sitting either side of the centre (~49)
        return pd.DataFrame({"x": [44.0, 54.0], "y": [49.0, 49.0],
                             "z": [49.0, 49.0], "mass": [1e8, 1e8]})
    # Swirling gas disc around the two black holes
    rng = np.random.default_rng(7)
    n = 70000
    centres = np.array([[44.0, 49.0], [54.0, 49.0]])
    which = rng.integers(0, 2, n)
    r = rng.exponential(2.5, n)
    th = rng.uniform(0, 2 * np.pi, n)
    x = centres[which, 0] + r * np.cos(th)
    y = centres[which, 1] + r * np.sin(th)
    z = 49.0 + rng.normal(0, 0.3, n)
    mass = rng.exponential(1e3, n)
    return pd.DataFrame({"x": x, "y": y, "z": z, "mass": mass})

# Load the two black holes and their surrounding gas disc
smbh_data = load_smbh("smbhs")     # columns: x, y, z, mass
gas_disc_data = load_smbh("gas")   # columns: x, y, z, mass

print("Black holes:")
print(smbh_data)
print("\nGas disc particles:", len(gas_disc_data))
```
