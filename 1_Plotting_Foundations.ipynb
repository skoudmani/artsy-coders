# Section 1: Hello Matplotlib! 📊

Let's begin with a basic **line plot**. A line plot connects data points which is perfect for plotting relationships in your data or functions.


```python
# Each code cell is self-contained: it imports what it needs.
%matplotlib inline
import matplotlib.pyplot as plt

# Anything after a '#' is a comment and is ignored by Python.
x = [0, 1, 2, 3, 4, 5, 6]
y = [0, 1, 4, 9, 16, 25, 36]

plt.plot(x, y)
plt.title("Simple Line Plot")
plt.xlabel("X values")
plt.ylabel("Y = X squared")
plt.grid(True)
plt.show()
```

# Section 2: Plotting Points (Scatter Plots) ✨

Sometimes we just want to plot individual **points** rather than a continuous line. This is where a **scatter plot** comes in!

Scatter plots are great for showing how two sets of values relate and to discover trends. Astronomers use these to show the positions of stars, or brightness vs. distance. Let's try `plt.scatter()`.


```python
%matplotlib inline
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [5, 3, 4, 7, 6]

plt.scatter(x, y)
plt.title("Scatter Plot Example")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

# Section 3: Adding Colour and Style 🎨🌟

Let's make our plots look **cooler**! We can change the **colour**, **size**, and **style** of our points.

In the example below we give each point its own colour and make the dots bigger with the `s` parameter (`s` is for "size").

👉 Pick your own colours from this list: [Matplotlib Colour Gallery](https://matplotlib.org/stable/gallery/color/named_colors.html)


```python
%matplotlib inline
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [5, 3, 4, 7, 6]
colors = ['red', 'green', 'blue', 'purple', 'orange']

plt.scatter(x, y, color=colors, s=100)  # 's' is the size of the dots
plt.title("Coloured Scatter Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
```

# Section 4: Plotting Multiple Lines 📈📉

What if we want to **compare** different patterns on the same graph? We can plot **multiple lines** at once.

Below we've plotted:
- `y = x` (a straight line)
- `y = x²` (a curve that grows quickly)
- `y = √x` (a curve that grows slowly)

🧠 **Challenge for you:**
Can you add another line for **`y = x³`**?
*Hint:* use `x**3`, add another `plt.plot(...)` line, and give it a label like `"y = x^3"` so it appears in the legend. You've got this! 🚀


```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 5, 100)
y1 = x
y2 = x**2
y3 = np.sqrt(x)

plt.figure(figsize=(6, 6))
plt.plot(x, y1, label="y = x")
plt.plot(x, y2, label="y = x^2")
plt.plot(x, y3, label="y = √x")
plt.title("Multiple Line Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
plt.xlim(0, 5)
plt.show()
```
